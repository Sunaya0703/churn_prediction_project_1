```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder , OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
from imblearn.over_sampling import RandomOverSampler
from sklearn.ensemble import RandomForestClassifier 
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split, cross_val_score 
```

```python
# Reading the dataset 

df=pd.read_csv('churn.csv')
df
```

```python
df.info()
```

```python
df.describe().T
```

```python
df1 = df.copy()
df1["day.charge"] = df1["day.charge"].astype("float64")
df1["eve.mins"]= df1["eve.mins"].astype("float64")
```

```python
df1.isnull().sum()
```

```python
df1.dropna(inplace = True)
```

```python
df1.drop(df1.columns[[0, 1, 2, 3]], axis=1, inplace=True)
df1
```

```python
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()
df1['churn']=LE.fit_transform(df1['churn'])
```

```python
df1=pd.get_dummies(df1,columns=['voice.plan','intl.plan'], drop_first=True)
df1
```

```python
df1['churn'].value_counts()
```

```python
X= df1.drop('churn', axis =1)
Y= df1.churn
```

```python
X
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify=Y, random_state=2)
```

```python
RS=RandomOverSampler()
X_train_rs, Y_train_rs = RS.fit_resample(X_train,Y_train)
```

```python
print(X.shape, X_train_rs.shape, X_test.shape)
```

```python
print(Y.shape,Y_train_rs.shape,Y_test.shape)
```

```python
RFC=RandomForestClassifier()
```

```python
# training the Random Forest Cassifier 


model=RFC.fit(X_train_rs,Y_train_rs)
```

```python
# accuracy score on the training data


X_train_prediction = RFC.predict(X_train_rs)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train_rs)
```

```python
print('Accuracy score of the training data : ', training_data_accuracy)
```

```python
# accuracy score on the test data


X_test_prediction = RFC.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score of the training data : ', test_data_accuracy)
```
